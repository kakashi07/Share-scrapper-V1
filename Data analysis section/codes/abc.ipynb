```python
import talib as ta 
import plotly_express as pe 
import pandas as pd 
import plotly.graph_objects as go 

filename = 'Share_Data_ADBL.csv'
data = pd.read_csv(filename, sep='|', header=0)
data['Date'] = pd.to_datetime(data.Date)
print(data.dtypes)
data = data.sort_values(by=['Date'])
# data.head()
SMA50 = (ta.SMA(data['ClosePrice'],timeperiod = 12)).fillna(0)
data['SMA50'] = SMA50
SMA200 = (ta.SMA(data['ClosePrice'],timeperiod = 26)).fillna(0)
data['SMA200'] = SMA200

RSI200 = (ta.RSI(data['ClosePrice'],timeperiod = 14)).fillna(0)
data['RSI14'] = RSI200/100
# print(data['Date'],data['RSI14'])
# fig = pe.line(data,x='Date',y='ClosePrice')
# fig.add_scatter(x=data['Date'],y=data['SMA50'])
# fig.update_layout(showlegend  = True)
# fig.show()
```

```python
fig = go.Figure()
rsi_figure = go.Figure()
fig.add_trace(go.Scatter(x=data['Date'], y=data['ClosePrice'],mode='lines',name='CP'))
fig.add_trace(go.Scatter(x=data['Date'], y=data['SMA50'],mode='lines',name='SMA50'))
fig.add_trace(go.Scatter(x=data['Date'], y=data['SMA200'],mode='lines',name='SMA200'))

rsi_figure.add_trace(go.Scatter(x=data['Date'], y=data['RSI14'],mode='lines',name='CP'))

fig.show()
rsi_figure.show()
```

```python
import datetime as dt
review_period =str(dt.date.today() - dt.timedelta(days = 5))
print(review_period)
latest_data = data[data['Date'] > review_period]

level1_flag1 = latest_data[(latest_data.SMA50 >= latest_data.SMA200)]
level2_flag1 = latest_data[latest_data.RSI14 < 0.40]

if len(level1_flag1.index)>0:
    level1_flag = 1
else:
    level1_flag = 0

if len(level2_flag1.index)>0:
    level2_flag = 1
else:
    level2_flag = 0

print(level1_flag,level2_flag)

x = ['Level1','Level2']
flag_figure = go.Figure([go.Bar(x=x,y=[level1_flag,level2_flag])])

flag_figure.show()
```
